**3. Are there some credit applicants that maybe considered as outliers? What makes them outliers?**

```python
import pandas as pd

# Import the data
application_data = pd.read_csv("application_record.csv")
credit_data = pd.read_csv("credit_record.csv")

# Merge the datasets on the 'ID' column
merged_data = pd.merge(application_data, credit_data, on='ID', how='left')

# Identify outliers based on credit record status
status_outliers = merged_data[merged_data["STATUS"].isin(["3", "4", "5"])]

# Print the identified outliers
print("Credit Record Status Outliers:")
print(status_outliers)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Import the data
application_data = pd.read_csv("application_record.csv")
credit_data = pd.read_csv("credit_record.csv")

# Merge the datasets on the 'ID' column
merged_data = pd.merge(application_data, credit_data, on='ID', how='left')

# Identify outliers based on credit record status
status_outliers = merged_data[merged_data["STATUS"].isin(["3", "4", "5"])]

# Count the occurrences of each credit record status
status_counts = status_outliers["STATUS"].value_counts()

# Define the descriptive labels for the credit record statuses
status_labels = {
    "3": "90-119 days overdue",
    "4": "120-149 days overdue",
    "5": "Overdue or bad debts,\nwrite-offs for more than 150 days"
}

# Update the x-axis labels with the descriptive labels
x_labels = [status_labels.get(status, status) for status in status_counts.index]

# Plot the identified outliers with descriptive x-axis labels
plt.figure(figsize=(8, 6))
plt.bar(x_labels, status_counts.values)
plt.xlabel("Credit Record Status")
plt.ylabel("Count")
plt.title("Credit Record Status Outliers")
plt.show()
```

<h4>summarize of the findings :</h4>

The analysis identified outliers in the credit record status, specifically focusing on statuses "3", "4", and "5". These statuses indicate varying degrees of overdue payments or bad debts. The descriptive labels for these statuses are as follows:

- <li>"3": 90-119 days overdue</li>
- <li>"4": 120-149 days overdue</li>
- <li>"5": Overdue or bad debts, write-offs for more than 150 days</li>

Upon analyzing the data, it was observed that the majority of the outliers fall into the "5" category, which represents cases of significant overdue payments or bad debts that have been written off for more than 150 days. This suggests a potential risk in lending to applicants who have a history of such credit record statuses.

By visualizing the outliers, we can gain insight into the prevalence of these problematic credit record statuses. The bar plot provides a clear representation of the count of outliers in each status category, with the x-axis displaying the descriptive labels for better understanding.

Overall, the analysis highlights the importance of monitoring and evaluating credit record status as a significant indicator of credit risk. The presence of a higher number of outliers in the "5" category signals the need for caution when considering lending decisions for applicants with such credit histories. This information can guide the bank's approach to lending, helping to mitigate potential risks and optimize decision-making processes.
