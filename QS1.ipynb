**1. Based on the provided dataset, perform some descriptive analysis about the relationship between annual income and education level?**

```python
import pandas as pd

# Import the data
data = pd.read_csv("application_record.csv")

# Calculate the statistics for annual income by education level
income_stats = data.groupby("NAME_EDUCATION_TYPE")["AMT_INCOME_TOTAL"].describe()

# Print the statistics
print(income_stats)
```

The code above calculates descriptive statistics (count, mean, standard deviation, min, quartiles, max) for annual income based on education level and prints the results. This provides a comprehensive understanding of the income distribution within each education level.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Import the data
data = pd.read_csv("application_record.csv")

# Group the data by education level
grouped_data = data.groupby("NAME_EDUCATION_TYPE")

# Calculate the mean annual income for each group
mean_income = grouped_data["AMT_INCOME_TOTAL"].mean()

# Create a table showing the mean annual income for each education level
table = pd.DataFrame(mean_income).reset_index()

# Visualize the data using a bar chart
plt.bar(table["NAME_EDUCATION_TYPE"], table["AMT_INCOME_TOTAL"])
plt.xlabel("Education Level")
plt.ylabel("Mean Annual Income")
plt.title("Mean Annual Income by Education Level")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

This code will generate a bar chart showing the mean annual income for each education level, with properly labeled axes and a rotated x-axis tick labels to prevent overlap.

<h2>Summary</h2>
<p> there is a positive correlation between annual income and education level. This means that people with higher education levels tend to have higher annual incomes.</p>
