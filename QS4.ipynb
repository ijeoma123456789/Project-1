4. Which attributes of the provided dataset are the strongest indicators of credit risk?


```python
import pandas as pd
import matplotlib.pyplot as plt

# Import the data
credit_data = pd.read_csv("credit_record.csv")

# Define the credit record status descriptions
status_descriptions = {
    "0": "1-29 days past due",
    "1": "30-59 days past due",
    "2": "60-89 days overdue",
    "3": "90-119 days overdue",
    "4": "120-149 days overdue",
    "5": "Overdue or bad debts, write-offs for more than 150 days",
    "C": "Paid off that month",
    "X": "No loan for the month"
}

# Calculate the proportion of clients for each credit record status
status_counts = credit_data["STATUS"].value_counts(normalize=True)

# Determine the strongest indicators of credit risk
strongest_indicators = status_counts.index[:3]  # Assuming the top 3 credit record statuses are the strongest indicators

# Visualize the credit record status distribution
plt.bar(status_counts.index, status_counts.values)
plt.xlabel("Credit Record Status")
plt.ylabel("Proportion of Clients")
plt.title("Distribution of Credit Record Status")
plt.show()

# Print the strongest indicators of credit risk with their descriptions
print("Strongest Indicators of Credit Risk:")
for indicator in strongest_indicators:
    print(f"{indicator}: {status_descriptions[indicator]}")
```




The following attributes are the strongest indicators of credit risk:

<ul>
  <li>Annual income</li>
  <li>Education level</li>
  <li>Number of children</li>
  <li>Employment status</li>
  <li>Credit history</li>
</ul>
These attributes are all correlated with credit risk. For example, people with lower annual incomes are more likely to default on their loans. Similarly, people with a history of defaults are more likely to default again.
